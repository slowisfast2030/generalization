```python
# 读取对测试集的预测。测试集的规模13586。
model_label_pred_label = pd.read_pickle('../generalization_data/sample_for_check_test_data_13586.pkl')

model_label_pred_label
```

```python

from sklearn import metrics

real_label = model_label_pred_label['real_label']
pred_label = model_label_pred_label['pred_label']

print('accuracy: ', metrics.accuracy_score(real_label, pred_label))
print('precision: ', metrics.precision_score(real_label, pred_label))
print('recall: ', metrics.recall_score(real_label, pred_label))
```

```python
# 读取20220831的所有数据并去重
all_data = pd.read_csv('../data_20220831/raw_cvjd_20220831_spark.csv').drop(['Unnamed: 0'], axis=1).drop_duplicates(subset=['cv_id', 'jd_id'], keep='first')
```

```python
all_data = all_data.reset_index(drop=True)
```

```python
# 适合抽样检查的jd列
jd_columns = ['title', 'category_name', 'company_name', 'recruit_number', 'work_age',
            'tags', 'workplaces', 'description', 'requirement', 'school',
            'min_annual_salary', 'max_annual_salary', 'min_month_salary', 'max_month_salary', 'pay_months']
all_data_jd = all_data[['cv_id', 'jd_id']+jd_columns]
```

```python
# 适合抽样检查的cv列
cv_columns = ['age', 'gender', 'certificates', 'currentCompany', 'currentPosition', 'desiredPosition',
            'currentSalary', 'desiredSalary', 'degree', 'industry', 'desiredIndustry', 'location', 'expectLocation', 'jobIntentStatus',
            'languageSkills', 'majorName', 'schoolName', 'skills', 'eduTracks', 'jobTracks', 'projectTracks', 'additionInfo', 'workYear']
all_data_cv = all_data[['cv_id', 'jd_id'] + cv_columns]
```

```python
# model_label_pred_label 
# all_data_jd 
# all_data_jd 
model_label_pred_label_jd = model_label_pred_label.join(all_data_jd.set_index(['cv_id', 'jd_id']), on=['cv_id', 'jd_id'], how='inner')
model_label_pred_label_jd_cv = model_label_pred_label_jd.join(all_data_cv.set_index(['cv_id', 'jd_id']), on=['cv_id', 'jd_id'], how='inner')
```
