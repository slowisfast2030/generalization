```python
# 拉取ai_prod.ads_ai_raw_label_a数据到generalization_data文件夹

import pandas as pd
import numpy as np

# 读取截止到20220908号的所有订单时间数据，规模87118
order_time_df = pd.read_csv('../generalization_data/raw_label_order_time_20220908_spark.csv').drop('Unnamed: 0', axis=1)
order_time_df
```

```python
# 以cv_id和jd_id进行去重，去重后规模83468
order_time_df_drop_dup = order_time_df.drop_duplicates(subset=['cv_id', 'jd_id'], keep='first').reset_index(drop=True)
order_time_df_drop_dup
```

```python
# 读取20220831的所有数据，规模83301
data_20220831 = pd.read_csv('../data_20220831/raw_cvjd_20220831_spark.csv').drop(['Unnamed: 0'], axis=1)
data_20220831
```

```python
# 以cv_id和jd_id进行去重，去重后规模82333
data_20220831_drop_dup = data_20220831.drop_duplicates(subset=['cv_id', 'jd_id'], keep='first').reset_index(drop=True)
data_20220831_drop_dup
```

```python
# order_time_df_drop_dup和data_20220831_drop_dup以['cv_id', 'jd_id']进行inner join
data_20220831_drop_dup_order_time = data_20220831_drop_dup.join(order_time_df_drop_dup.set_index(['cv_id', 'jd_id']), on=['cv_id', 'jd_id'], how='inner')
data_20220831_drop_dup_order_time[['cv_id', 'jd_id', 'pair_created_time']]
```

```python
cv_jd_time = data_20220831_drop_dup_order_time[['cv_id', 'jd_id', 'pair_created_time']].sort_values(by='pair_created_time').reset_index(drop=True)
cv_jd_time
```

```python
# pair_created_time列只保留到天
def get_day_fun(series):
    pair_created_time = series['pair_created_time']
    day_time = pair_created_time.split(' ')[0]
    return day_time

# pair_created_time列只保留到月
def get_month_fun(series):
    pair_created_time = series['pair_created_time']
    day_time = pair_created_time.split(' ')[0]
    return day_time[:7]

cv_jd_time['day'] = cv_jd_time.apply(get_day_fun, axis=1)
cv_jd_time['month'] = cv_jd_time.apply(get_month_fun, axis=1)
cv_jd_time
```

```python
# cv_jd_time表的数据需要保存
cv_jd_time.to_csv('../generalization_data/cv_jd_time_20220831_82333.pkl')
```
