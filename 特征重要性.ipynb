```python
import tensorflow as tf

# importing graph

saver = tf.train.import_meta_graph('../modelV0915/model.ckpt-1000.meta')
```

```python
import numpy as np
import pandas as pd
```

```python
with tf.Session() as sess:
    sess.run(tf.tables_initializer())

# loading variable value to sess

    saver.restore(sess, "../modelV0915/model.ckpt-1000")

    result =  tf.get_default_graph().get_tensor_by_name("add:0")   

    print(sess.run(result)) 
    
    variable_names = [v.name for v in tf.trainable_variables()]
    values = sess.run(variable_names)
    for i,j in zip(variable_names, values):
        print(i)
        print(j)
```
